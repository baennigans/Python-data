```python
print("1번-----------------------------------------------------------------")
import seaborn as sns
flights = sns.load_dataset('flights')
print("flights 데이터의 shape : ", flights.shape,"\n")
print(flights.head())
```

```python
print("2번-----------------------------------------------------------------")
import pandas as pd
yearly_passengers = flights.groupby('year')['passengers'].sum()
print(yearly_passengers)
```

```python
print("3번-----------------------------------------------------------------")
import matplotlib.pyplot as plt
sns.lineplot(x=yearly_passengers.index, y=yearly_passengers.values, label='passengers')
```

```python
print("4번-----------------------------------------------------------------")
plt.scatter(yearly_passengers.index, yearly_passengers.values, marker='o', s=10, c='blue')
```

```python
print("\n5번-----------------------------------------------------------------")
from sklearn import linear_model
from sklearn.metrics import r2_score

regr = linear_model.LinearRegression()
data = pd.DataFrame({'year': yearly_passengers.index, 'passengers': yearly_passengers.values})
X = data[['year']]
y = data['passengers']

regr.fit(X, y)
y_pred = regr.predict(X)
score = r2_score(y, y_pred)
print(f"선형 회귀 모델의 계수 : {regr.coef_}")
print(f"선형 회귀 모델의 절편 : {regr.intercept_}")
print(f"데이터와 선형회귀 직선의 관계점수 : {score} %\n\n")
```

```python
print("6번-----------------------------------------------------------------")
import numpy as np
plt.scatter(X, y, marker='o', s=10, c='blue')
x_range = np.arange(X.min()[0], X.max()[0] + 1).reshape(-1, 1)
y_pred = regr.predict(x_range)
plt.plot(x_range, y_pred, color='red', linewidth=2)
```
