```python
# 유임승차, 유임하차, 무임승차, 무임하차 승객수가 가장 많은 역과 승객수 조회
import csv
import matplotlib.pyplot as plt
```

```python
with open('data/202308_subway.csv', 'r', encoding='utf-8') as file:
    data = csv.reader(file)
    next(data)
    # maxInfo[0] : 유임승차 승객수가 가장 많은 역
    # maxInfo[1] : 유임승차 승객수가 가장 많은 역의 승객수 
    maxInfo = ['', 0]  
    for row in data:
        row[4] = int(row[4].replace(',',''))
        if row[4] > maxInfo[1]:
            maxInfo[0] = row[3]
            maxInfo[1] = row[4]
    print(f'유임승차 승객수가 가장 많은 역은 {maxInfo[0]}이고, 승객수는 {maxInfo[1]}명이다')
```

```python
# 유임승차, 유임하차, 무임승차, 무임하차 승객수가 
# 가장 많은 역과 승객수 조회
with open('data/202308_subway.csv', 'r', encoding='utf-8') as file:
    data = csv.reader(file)
    next(data)
    stations = [''] * 4
    max_persons = [0] * 4
    # print(stations, max_persons)
    for row in data:
        for i in range(4):
            person = int(row[4+i].replace(',',''))
            if person > max_persons[i]:
                stations[i] = row[3]
                max_persons[i] = person
        
    titles = ['유임승차', '유임하차', '무임승차', '무임하차']
    for i in range(len(stations)):
        print(f'{titles[i]} : {stations[i]}({max_persons[i]}명)')
```

```python
# 출근시간(오전7시)에 가장 많이 승객이 승차하는 역을 검색
with open('data/202308_trans_time.csv', 'r') as file:
    data = csv.reader(file)
    next(data)
    next(data)
    
    person_list = []
    station = ''
    max_person = 0
    for row in data:
        person = int(row[10].replace(',',''))
        person_list.append(person)
        if person > max_person:
            station = f'{row[3]}({row[1]})'
            max_person = person
    print(f'{station} : {max_person}명')
    
    plt.plot(range(len(person_list)), person_list)
```

```python
# 출근시간대(7시, 8시, 9시) 승차승객수가 가장 많은 역 조회

def to_int(data):
    return int(data.replace(',', ''))

with open('data/202308_trans_time.csv', 'r') as file:
    data = csv.reader(file)
    next(data)
    next(data)
    max_persons = 0

    station = ''
    for row in data:
        persons = row[10:15:2]
        persons = list(map(to_int, persons))
        # [to_int(row[10]), to_int(row[12]), to_int(row[14])]
        
        #persons = [int(p.replace(',', '')) for p in row[10:15:2]]
        #print(persons, sum(persons))
        
        if sum(persons) > max_persons:
            station = f'{row[3]}({row[1]})'
            max_persons = sum(persons)
            
    print(f'{station} : {max_persons}명')
```

```python
# 출근시간대(7시, 8시, 9시) 승차승객수가 가장 많은 역 조회

def to_int(data):
    return int(data.replace(',', ''))

with open('data/202308_trans_time.csv', 'r') as file:
    data = csv.reader(file)
    next(data)
    next(data)
    max_persons = [0]
    station = ''
    for row in data:
        persons = row[10:15:2]
        persons = list(map(to_int, persons))
        # [to_int(row[10]), to_int(row[12]), to_int(row[14])]
        
        #persons = [int(p.replace(',', '')) for p in row[10:15:2]]
        #print(persons, sum(persons))
        
        if sum(persons) > sum(max_persons):
            station = f'{row[3]}({row[1]})'
            max_persons = persons
            
    print(f'{station} : {sum(max_persons)}명')
  
    plt.rc('font', family='gulim')
    plt.pie(max_persons, labels=['7시대', '8시대', '9시대']
            , autopct='%.2f%%')
```

```python
# 검색할 시간대를 입력(24시) : 9
# 9시대에 가장 많은 승차인원을 가진 역과 인원수 출력

with open('data/202308_trans_time.csv', 'r') as file:
    data = csv.reader(file)
    next(data)
    next(data)
    
    inputTime = int(input('검색할 시간대를 입력하세요(0-24) : '))
    
    station = ''
    max_persons = 0
    index = inputTime * 2 - 4
    
    for row in data:
        row[index] = row[index].replace(',', '')
        if max_persons < int(row[index]) :
            max_persons = int(row[index])
            station = f'{row[3]}({row[1]})'
    
    print(f'{inputTime}시대 : {station}[{max_persons}명]')
```
