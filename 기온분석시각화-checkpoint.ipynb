```python
with open('data/seoul-2022.csv', 'r') as file:
    data = csv.reader(file)
    highTemplist = list()
    next(data)
    for row in data:
        # print(row[-1])
        #print(f'[{row}]')
        if len(row) > 1 and row[-1] != '':
            highTemplist.append(float(row[-1]))
    plt.figure(figsize=(10, 3))
    plt.plot(highTemplist, 'r')
```

```python
# 매년 8월의 최고온도데이터를 시각화
with open('data/seoul-1970and2023.csv', 'r') as file:
    data = csv.reader(file)
    next(data)
    highTemplist = []
    for row in data:
        if len(row) > 1 and row[-1] != '':
            date = row[0].split('-')
            month = date[1]
            if month == '08':
                # print(row)
                highTemplist.append(float(row[-1]))
    
    plt.figure(figsize=(10,3))
    plt.plot(highTemplist, 'r')
```

```python
# 자신의 생일의 기온그래프(최고, 최저) 시각화
with open('data/seoul-1970and2023.csv', 'r') as file:
    highTemp = []
    lowTemp = []
    years = []
    
    data = csv.reader(file)
    next(data)
    for row in data:
        if len(row) > 1 and row[-1] != '':
            ymd = row[0].split('-')
            if ymd[1] == '10' and ymd[2] == '21':
                #print(row)
                highTemp.append(float(row[-1]))
                lowTemp.append(float(row[-2]))
                years.append(ymd[0].strip())
    #print(highTemp)
    #print(lowTemp)
    
    plt.figure(figsize=(15, 3))
    plt.xticks(rotation=90)
    plt.plot(years, highTemp, 'r')
    plt.plot(years, lowTemp, 'b')
```
